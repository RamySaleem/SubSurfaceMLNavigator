## Introduction

The patterns of the displacement-distance profiles were originally derived from displacement variations vertically along the thrust trace, and the distance from the reference point to the tips of the thrust, wherein any large thrust displacement reduces towards the two ends of the thrust trace.

More specifically, the recognition that the displacement-distance profiles were at a maximum near the centre and diminished outwards in all directions to the edge of the thrust where the displacement is zero led to the understanding that the displacement-distance profile permanently had a simple shape of normal distribution in the idealised model. 

More recent observations, primarily from regional and outcrop field studies and seismic reflection data, have supported an alternative, more complex displacement profile pattern. 

In this, the displacement profiles show variations from the normal distribution by increasing or decreasing the displacement in which the displacement-distance profiles are possibly controlled by mechanical stratigraphy.

However, another explanation could be inaccurate interpretations in which the relation between faults and horizons is misinterpreted, resulting in abnormal fold-thrust geometries and displacement distribution.

 A third, more kinematic explanation posits that the displacement is transferred either by fault to fault interaction (hard linkage) or fault to fold interaction (soft linkage) which also resulting in a complex displacement profile pattern.

## Linear Regression using a Single Feature (Displacement)

We now know that the "Displacement" and "Length" columns have the strongest correlation alongside with "distance" and "profile patterns". Let's try to find a way of estimating the value of "Length" using the value of "Displacement" for faults. First, let's create a data frame containing just the data for displacement.

Apart from a few exceptions, the points seem to form a weak trend. We'll try and "fit" a line using this points, and use the line to predict length for a given dispalcement. A line on the X&Y coordinates has the following formula:

$y = wx + b$

The line is characterized two numbers: $w$ (called "slope") and $b$ (called "intercept"). 

### Model

In the above case, the x axis shows "dispalcement" and the y axis shows "length". Thus, we're assume the following relationship between the two:

$Length = w \times dispalcement + b$

We'll try determine $w$ and $b$ for the line that best fits the data. 

* This technique is called _linear regression_, and we call the above equation a _linear regression model_, because it models the relationship between "Max_Displacement" and "Length" as a straight line. 

* The numbers $w$ and $b$ are called the _parameters_ or _weights_ of the model.

* The values in the "Dispalcement" column of the dataset are called the _inputs_ to the model and the values in the "Length" column are called "targets". 

Let define a helper function `estimate_Max_Displacement`, to compute $length$, given $dispalcement$, $w$ and $b$.

```python
def estimate_Max_Displacement(length, w, b):
    return w * length + b
```

```python
def test_params(**params):
    model = RandomForestRegressor(random_state=42, n_jobs=-1, **params).fit(inputs_train, targets_train)
    train_rmse = mean_squared_error(model.predict(inputs_train), targets_train, squared=False)
    test_rmse = mean_squared_error(model.predict(inputs_test), targets_test, squared=False)
    return train_rmse, test_rmse
```

```python
test_params(n_estimators=20, max_depth=25)
```

```python
test_params(n_estimators=2, max_depth=14, min_samples_leaf=4, max_features=0.4)
```

```python
def test_param_and_plot(param_name, param_values):
    train_errors, val_errors = [], [] 
    for value in param_values:
        params = {param_name: value}
        train_rmse, val_rmse = test_params(**params)
        train_errors.append(train_rmse)
        val_errors.append(val_rmse)
    plt.figure(figsize=(10,6))
    plt.title('Overfitting curve: ' + param_name, fontsize=18)
    plt.plot(param_values, train_errors, 'b-o')
    plt.plot(param_values, val_errors, 'r-o')
    plt.xlabel(param_name, fontsize=18)
    plt.ylabel('RMSE', fontsize=18)
    plt.legend(['Training', 'Validation'], fontsize=18)
```
